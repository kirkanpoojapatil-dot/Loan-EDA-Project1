```python
def detecting_outliers_with_IQR(data, factor):
    IQR=np.quantile(data, 0.75)-np.quantile(data, 0.25)
    LL=np.quantile(data, 0.25)-(factor*IQR)
    UL=np.quantile(data, 0.25)+(factor*IQR)
    return LL,UL
```

```python
data=df["ApplicantIncome"]
LL,UL=detecting_outliers_with_IQR(data, 3)
```

```python
LL
```

```python
UL
```

```python
def outlier_detection_using_quantile(data):
    LL1=np.quantile(data,0.01)
    UL1=np.quantile(data,0.99) 
    return LL1,UL1
```

```python
data=df["ApplicantIncome"]
LL,UL=outlier_detection_using_quantile(data)
```
